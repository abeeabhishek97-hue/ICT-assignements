```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)


def sigmoid(x):
    return 1 / (1 + np.exp(-x))
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.plot(x, sigmoid(x))
plt.title("Sigmoid")
plt.grid()
```

```python
def linear(x):
  return x
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 2)
plt.plot(x, linear(x))
plt.title("linear")
plt.grid()
```

```python
def relu(x):
  return np.maximum(0,x)
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 3)
plt.plot(x, relu(x))
plt.title("relu")
plt.grid()
```

```python
def tanh(x):
  return np.tanh(x)
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 4)
plt.plot(x, tanh(x))
plt.title("tanh")
plt.grid()
```

```python
def softmax(x):
  return np.exp(x) / np.sum(np.exp(x))
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 5)
plt.plot(x, softmax(x))
plt.title("softmax")
plt.grid()
```
